```python
# scatterplot-- is draw between two numerical columns
import pandas as pd
iris=pd.read_csv("iris.csv")
iris.head()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
sns.scatterplot(x="Sepal Length (cm)",y="Petal Length (cm)",data=iris,hue="Class",style="Class")
plt.show()
```

```python
# petal length value increase, color intensityis also increased
sns.scatterplot(x="Sepal Length (cm)",y="Petal Length (cm)",data=iris,hue="Petal Length (cm)")
plt.show()
```

```python
# seaborn histogram/Distplot-- histogram and frequency curve combination
# single numerical distribution check--use histogram
sns.distplot(iris['Petal Width (cm)'])
plt.show()
```

```python
# don't show the histogram
sns.distplot(iris['Petal Width (cm)'],hist=False)
plt.show()
```

```python
sns.distplot(iris['Petal Width (cm)'],color="green")
plt.show()
```

```python
# no need of frequency curve then kde=false
sns.distplot(iris['Petal Width (cm)'],color="green",bins=4,kde=False)
plt.show()
```

```python
# distplot is used to analyze 1 feature(univariate analysis)
# joinplot is used for 2 features(bivariate analysis)
#pairplot is used for more than 2 features
```

```python
# inbuilt dataset which is load_dataset
df=sns.load_dataset("tips")
df.head()
#Tip --dependent feature and all others are independent features
```

```python
df.dtypes
```

```python
#correalation is used to find the inter-relationship between all the features
# corealtion is find out when the values are either int or float.
#If values are categorial(object)then we didn't find the inter-realtionship 
# value of corelation lies betwwn -1 to +1
#There are 2 main types of categorical data, 
#namely; nominal data and ordinal data. 
#Nominal Data--that have names or labels. Sometimes called naming data
#Eg. Name of a person,  gender, school graduates from,  etc
# Ordinal Data--includes elements that are ranked, ordered 
```

```python
df.corr()
```

```python
#heatmap is used to check the visualization
# input--matrix of correation
sns.heatmap(df.corr())
#highly corelated is given to 1 value and that color , check the color variance
```

```python
sns.jointplot(x='tip',y='total_bill',data=df)
```

```python
# kind means show all the features in hexagonal shape
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
# reg draw probability density function and regression line
```

```python
# pairplot is used --if we have more than 2 features
sns.pairplot(df)
```
